The Wesenheit index is defined as `W = I - 1.55(V - I)`, and its main advantage over using simply the I or V photometry is that it is insensitive to extinction. It is denoted by 'W' among the data columns.

## Part 3: Compute the residuals of each sample to its respective line.
Do these residuals look like a normal distribution? If not, speculate on why (WATCH THE YOUTUBE VIDEO!)

Yes, the distribution of the residuals is normal and indicates that the errors between the observed values and the predicted values follow a symmetry (with values mostly close to 0). This suggests that the regression model used to estimate the period-luminosity relationship of Cepheid stars is appropriate.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

data = pd.read_csv('Cepheids.csv')

lmc_fundamental = data[(data['Cloud'] == 'LMC') & (data['Mode'] == 'F')]
lmc_overtone = data[(data['Cloud'] == 'LMC') & (data['Mode'] == '1')]
smc_fundamental = data[(data['Cloud'] == 'SMC') & (data['Mode'] == 'F')]
smc_overtone = data[(data['Cloud'] == 'SMC') & (data['Mode'] == '1')]

#Residuals for LMC Fundamental-mode Cepheids
slope, intercept, r_value, p_value, std_err = stats.linregress(lmc_fundamental['logP1'], lmc_fundamental['W'])
residuals_lmc_fundamental = lmc_fundamental['W'] - (intercept + slope * lmc_fundamental['logP1'])

# Plotting residuals for LMC Fundamental-mode Cepheids
plt.figure(figsize=(10, 6))
plt.scatter(lmc_fundamental['logP1'], residuals_lmc_fundamental, color='blue')
plt.axhline(y=0, color='black', linestyle='--')
plt.xlabel('log(P1)')
plt.ylabel('Residuals')
plt.title('Residuals of Period-Luminosity Relationship for LMC Fundamental-mode Cepheids')
plt.grid(True)
plt.show()

#Residuals for LMC First overtone Cepheids
slope, intercept, r_value, p_value, std_err = stats.linregress(lmc_overtone['logP1'], lmc_overtone['W'])
residuals_lmc_overtone = lmc_overtone['W'] - (intercept + slope * lmc_overtone['logP1'])

# Plotting residuals for LMC First overtone Cepheids
plt.figure(figsize=(10, 6))
plt.scatter(lmc_overtone['logP1'], residuals_lmc_overtone, color='red')
plt.axhline(y=0, color='black', linestyle='--')
plt.xlabel('log(P1)')
plt.ylabel('Residuals')
plt.title('Residuals of Period-Luminosity Relationship for LMC First overtone Cepheids')
plt.grid(True)
plt.show()

#Residuals for SMC Fundamental-mode Cepheids
slope, intercept, r_value, p_value, std_err = stats.linregress(smc_fundamental['logP1'], smc_fundamental['W'])
residuals_smc_fundamental = smc_fundamental['W'] - (intercept + slope * smc_fundamental['logP1'])

# Plotting residuals for SMC Fundamental-mode Cepheids
plt.figure(figsize=(10, 6))
plt.scatter(smc_fundamental['logP1'], residuals_smc_fundamental, color='green')
plt.axhline(y=0, color='black', linestyle='--')
plt.xlabel('log(P1)')
plt.ylabel('Residuals')
plt.title('Residuals of Period-Luminosity Relationship for SMC Fundamental-mode Cepheids')
plt.grid(True)
plt.show()

#Residuals for SMC First overtone Cepheids
slope, intercept, r_value, p_value, std_err = stats.linregress(smc_overtone['logP1'], smc_overtone['W'])
residuals_smc_overtone = smc_overtone['W'] - (intercept + slope * smc_overtone['logP1'])

# Plotting residuals for SMC First overtone Cepheids
plt.figure(figsize=(10, 6))
plt.scatter(smc_overtone['logP1'], residuals_smc_overtone, color='orange')
plt.axhline(y=0, color='black', linestyle='--')
plt.xlabel('log(P1)')
plt.ylabel('Residuals')
plt.title('Residuals of Period-Luminosity Relationship for SMC First overtone Cepheids')
plt.grid(True)
plt.show()
```

## Part 4: Scatter plot of the residuals as RA (x-axis) vs Dec (y-axis) color-coded by whether they are positive or negative.
What do you see?

The residuals do not show a systematic tendency to deviate in any particular direction from the coordinates. Both positive and negative are together, therefore, it is assumed that the model is appropriate and the estimates obtained are reliable, since the errors between the observed values and the predicted values are randomly distributed and do not present a discernible pattern related to the coordinates.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

data = pd.read_csv('Cepheids.csv')

lmc_fundamental = data[(data['Cloud'] == 'LMC') & (data['Mode'] == 'F')]
lmc_overtone = data[(data['Cloud'] == 'LMC') & (data['Mode'] == '1')]
smc_fundamental = data[(data['Cloud'] == 'SMC') & (data['Mode'] == 'F')]
smc_overtone = data[(data['Cloud'] == 'SMC') & (data['Mode'] == '1')]

#Residuals for LMC Fundamental-mode Cepheids
slope, intercept, r_value, p_value, std_err = stats.linregress(lmc_fundamental['logP1'], lmc_fundamental['W'])
residuals_lmc_fundamental = lmc_fundamental['W'] - (intercept + slope * lmc_fundamental['logP1'])

# Scatter plot of residuals for LMC Fundamental-mode Cepheids
plt.figure(figsize=(10, 6))
positive_residuals_lmc_fundamental = lmc_fundamental[residuals_lmc_fundamental >= 0]
negative_residuals_lmc_fundamental = lmc_fundamental[residuals_lmc_fundamental < 0]

plt.scatter(positive_residuals_lmc_fundamental['RA0'], positive_residuals_lmc_fundamental['Decl0'], color='blue', label='Positive Residuals')
plt.scatter(negative_residuals_lmc_fundamental['RA0'], negative_residuals_lmc_fundamental['Decl0'], color='red', label='Negative Residuals')

plt.xlabel('RA')
plt.ylabel('Dec')
plt.title('Residuals of LMC Fundamental-mode Cepheids: RA vs Dec')
plt.legend()
plt.grid(True)
plt.show()

#Residuals for LMC First overtone Cepheids
slope, intercept, r_value, p_value, std_err = stats.linregress(lmc_overtone['logP1'], lmc_overtone['W'])
residuals_lmc_overtone = lmc_overtone['W'] - (intercept + slope * lmc_overtone['logP1'])

# Scatter plot of residuals for LMC First overtone Cepheids
plt.figure(figsize=(10, 6))
positive_residuals_lmc_overtone = lmc_overtone[residuals_lmc_overtone >= 0]
negative_residuals_lmc_overtone = lmc_overtone[residuals_lmc_overtone < 0]

plt.scatter(positive_residuals_lmc_overtone['RA0'], positive_residuals_lmc_overtone['Decl0'], color='blue', label='Positive Residuals')
plt.scatter(negative_residuals_lmc_overtone['RA0'], negative_residuals_lmc_overtone['Decl0'], color='red', label='Negative Residuals')

plt.xlabel('RA')
plt.ylabel('Dec')
plt.title('Residuals of LMC First overtone Cepheids: RA vs Dec')
plt.legend()
plt.grid(True)
plt.show()

#Residuals for SMC Fundamental-mode Cepheids
slope, intercept, r_value, p_value, std_err = stats.linregress(smc_fundamental['logP1'], smc_fundamental['W'])
residuals_smc_fundamental = smc_fundamental['W'] - (intercept + slope * smc_fundamental['logP1'])

# Scatter plot of residuals for SMC Fundamental-mode Cepheids
plt.figure(figsize=(10, 6))
positive_residuals_smc_fundamental = smc_fundamental[residuals_smc_fundamental >= 0]
negative_residuals_smc_fundamental = smc_fundamental[residuals_smc_fundamental < 0]

plt.scatter(positive_residuals_smc_fundamental['RA0'], positive_residuals_smc_fundamental['Decl0'], color='blue', label='Positive Residuals')
plt.scatter(negative_residuals_smc_fundamental['RA0'], negative_residuals_smc_fundamental['Decl0'], color='red', label='Negative Residuals')

plt.xlabel('RA')
plt.ylabel('Dec')
plt.title('Residuals of SMC Fundamental-mode Cepheids: RA vs Dec')
plt.legend()
plt.grid(True)
plt.show()

#Residuals for SMC First overtone Cepheids
slope, intercept, r_value, p_value, std_err = stats.linregress(smc_overtone['logP1'], smc_overtone['W'])
residuals_smc_overtone = smc_overtone['W'] - (intercept + slope * smc_overtone['logP1'])

# Scatter plot of residuals for SMC First overtone Cepheids
plt.figure(figsize=(10, 6))
positive_residuals_smc_overtone = smc_overtone[residuals_smc_overtone >= 0]
negative_residuals_smc_overtone = smc_overtone[residuals_smc_overtone < 0]

plt.scatter(positive_residuals_smc_overtone['RA0'], positive_residuals_smc_overtone['Decl0'], color='blue', label='Positive Residuals')
plt.scatter(negative_residuals_smc_overtone['RA0'], negative_residuals_smc_overtone['Decl0'], color='red', label='Negative Residuals')

plt.xlabel('RA')
plt.ylabel('Dec')
plt.title('Residuals of SMC First overtone Cepheids: RA vs Dec')
plt.legend()
plt.grid(True)
plt.show()
```
